```python
df["D"].hist(bins=100)  # "D" - specify column. #histogram of distrubution 
```

### Matplotlib

```python
import matplotlib.pyplot as plt 
```

```python
%matplotlib inline 
## Allows you to see the plots inside jupyter notebook
```

```python
y = [1,2,3,4]
```

```python
plt.plot(y)
```

```python
x = np.linspace(0,5,11) 
y = x**2
```

```python
fig = plt.figure() ## think of it as my black canvas
axes = fig.add_axes([0.1,0.1, 0.8, 0.8])
axes.plot(x,y)
axes.set_xlabel("X axis")
axes.set_title("this is the title")
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax1.plot(x,y)
ax2.plot(y,x)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
plt.tight_layout()
axes[0].plot(x,y)
axes[0].set_title("plot 1")
axes[1].plot(y,x)
axes[1].set_title("plot 2")
```
